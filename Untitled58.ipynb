## Compare Revenue & Budget Movies

```python
##IF revenue>Budget Movies Succsess 
##IF revenue<Budget Movies Fail
df['Larger Values']=df[['Budget','Revenue']].max(axis=1)
df['Smaller Values']=df[['Budget','Revenue']].min(axis=1)
```

```python
df['Movie Statue']=df['Revenue']>df['Budget']
```

```python
df['Movie Statue'].value_counts()
```
